## Introduction

### Motivation

In this section, we will develop an intuitive understanding of **backpropagation**, which is a method for computing gradients of expressions through the recursive application of the **chain rule**.

Understanding this process and its subtleties is critical for:

* Designing neural networks
* Developing models
* Debugging effectively

---

### Problem Statement

The core problem studied in this section is:

$$
f(x)
$$

where ( x ) is a vector of inputs, and we are interested in computing:

$$
\nabla f(x)
$$

which is the **gradient of the function** at ( x ).

---

### Motivation (Neural Networks Context)

In neural networks:

* ( f ) corresponds to the **loss function** ( L )
* ( x ) includes both **training data** and **model parameters**

Example:

* Training data:
  $$
  (x_i, y_i), \quad i = 1 \dots N
  $$

* Parameters:
  $$
  W, b
  $$

Key points:

* Training data → fixed (given)
* Parameters → variables we optimize

Therefore, in practice:

$$
\frac{\partial L}{\partial W}, \quad \frac{\partial L}{\partial b}
$$

are the gradients we compute to update the model.

However, gradients with respect to ( x_i ) can also be useful:

* Visualization
* Model interpretation

---

### Note

Even if you are already comfortable with the chain rule, it is recommended to skim this section.

It presents backpropagation as a:

👉 **backward flow in real-valued circuits**

which provides useful intuition.

---

## Simple Expressions and Interpretation of the Gradient

### Simple Example

Consider the function:

$$
f(x, y) = xy
$$

Partial derivatives:

$$
\frac{\partial f}{\partial x} = y
$$

$$
\frac{\partial f}{\partial y} = x
$$

---

### Interpretation of the Derivative

$$
\frac{df(x)}{dx} = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
$$

This measures how the function changes for an **infinitesimally small change** in input.

Linear approximation:

$$
f(x + h) \approx f(x) + h \frac{df(x)}{dx}
$$

---

### Intuition

The derivative represents **sensitivity**.

Example:

* ( x = 4 ), ( y = -3 )

$$
f(x, y) = -12
$$

$$
\frac{\partial f}{\partial x} = -3
$$

Interpretation:

* Increasing ( x ) slightly → decreases output
* Change is approximately **3 times larger**

---

### Gradient Definition

$$
\nabla f = \left[ \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right] = [y, x]
$$

---

## Addition Operation

$$
f(x, y) = x + y
$$

$$
\frac{\partial f}{\partial x} = 1
$$

$$
\frac{\partial f}{\partial y} = 1
$$

Interpretation:

* Increasing either variable increases the output
* Rate is constant and independent of values

---

## Max Operation

$$
f(x, y) = \max(x, y)
$$

$$
\frac{\partial f}{\partial x} =
\begin{cases}
1 & \text{if } x \ge y \
0 & \text{otherwise}
\end{cases}
$$

$$
\frac{\partial f}{\partial y} =
\begin{cases}
1 & \text{if } y \ge x \
0 & \text{otherwise}
\end{cases}
$$

---

### Intuition

Example:

* ( x = 4 ), ( y = 2 )

$$
f(x, y) = 4
$$

* Increasing ( y ) slightly → no effect
  👉 gradient = 0

Important:

* Derivatives only describe **infinitesimally small changes**

---

### Summary

The derivative with respect to each variable tells you the **sensitivity of the function** to that variable.


## Compound Expressions with Chain Rule

### Motivation

Now we consider more complex expressions composed of multiple functions:

$$
f(x, y, z) = (x + y)z
$$

Instead of differentiating directly, we break it into intermediate steps:

$$
q = x + y
$$

$$
f = qz
$$

---

### Local Derivatives

From previous rules:

For multiplication:

$$
\frac{\partial f}{\partial q} = z
$$

$$
\frac{\partial f}{\partial z} = q
$$

For addition:

$$
\frac{\partial q}{\partial x} = 1
$$

$$
\frac{\partial q}{\partial y} = 1
$$

---

### Chain Rule

We are interested in gradients with respect to inputs (x, y, z), not the intermediate variable (q).

Using the chain rule:

$$
\frac{\partial f}{\partial x} = \frac{\partial f}{\partial q} \cdot \frac{\partial q}{\partial x}
$$

$$
\frac{\partial f}{\partial y} = \frac{\partial f}{\partial q} \cdot \frac{\partial q}{\partial y}
$$

---

### Example (Forward + Backward Pass)

```python
# set some inputs
x = -2
y = 5
z = -4

# forward pass
q = x + y      # q = 3
f = q * z      # f = -12

# backward pass
dfdz = q       # = 3
dfdq = z       # = -4

dqdx = 1.0
dqdy = 1.0

dfdx = dfdq * dqdx
dfdy = dfdq * dqdy
```

Final gradients:

$$
[dfdx, dfdy, dfdz]
$$

These represent the **sensitivity of each input variable** on the output (f).

---

### Notation Simplification

Instead of writing:

$$
\frac{\partial f}{\partial q}
$$

we will use:

$$
dq
$$

and assume gradients are always taken with respect to the final output.

## Intuitive Understanding of Backpropagation

### Local Nature of Backpropagation

Backpropagation is a **local process**.

Each gate in a computational graph:

1. Computes its **forward output**
2. Computes its **local gradient** with respect to its inputs

Importantly:

* Each gate works **independently**
* It does **not need global knowledge** of the entire circuit

---

### Chain Rule in Backpropagation

During the backward pass:

* Each gate receives the gradient from the output
* Then **multiplies it with its local gradients**

This is the chain rule:

$$
\text{upstream gradient} \times \text{local gradient}
$$

---

### Key Insight

> A simple gate becomes powerful when combined with others through the chain rule.

---

### Intuition with Example

Recall:

$$
f(x, y, z) = (x + y)z
$$

Forward pass:

* Add gate: inputs ([-2, 5]) → output (3)
* Multiply gate → final output (-12)

Backward pass:

* Add gate receives gradient: (-4)

Interpretation:

* The system “wants” the add gate output to **decrease**
* Strength of this signal: **4**

Then:

$$
\frac{\partial f}{\partial x} = -4
$$

$$
\frac{\partial f}{\partial y} = -4
$$

---

### Intuition

If (x) and (y) decrease:

* Add gate output decreases
* Multiply output increases

---

### Big Picture

> Backpropagation is like gates **communicating signals**:
>
> * Increase output
> * Decrease output
> * How strongly

### Additional Gates

We introduce more functions:

$$
f(x) = \frac{1}{x} \quad \rightarrow \quad \frac{df}{dx} = -\frac{1}{x^2}
$$

$$
f_c(x) = x + c \quad \rightarrow \quad \frac{df}{dx} = 1
$$

$$
f(x) = e^x \quad \rightarrow \quad \frac{df}{dx} = e^x
$$

$$
f_a(x) = ax \quad \rightarrow \quad \frac{df}{dx} = a
$$

---

### Sigmoid Function

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

Derivative:

$$
\frac{d\sigma(x)}{dx} = (1 - \sigma(x)) \sigma(x)
$$

---

### Insight

The gradient simplifies to:

$$
(1 - \sigma(x)) \cdot \sigma(x)
$$

Example:

* Input: (1.0)
* Output: (0.73)

$$
(1 - 0.73) \cdot 0.73 \approx 0.2
$$

---

### Key Takeaway

> Complex expressions can be grouped into **single efficient gates**.


````python
import math
w = [2,-3,-3] # assume some random weights and data
x = [-1, -2]

# forward pass
dot = w[0]*x[0] + w[1]*x[1] + w[2]
f = 1.0 / (1 + math.exp(-dot)) # sigmoid function

# backward pass through the neuron (backpropagation)
ddot = (1 - f) * f # gradient on dot variable, using the sigmoid gradient derivation
dx = [w[0] * ddot, w[1] * ddot] # backprop into x
dw = [x[0] * ddot, x[1] * ddot, 1.0 * ddot] # backprop into w
# we're done! we have the gradients on the inputs to the circuit
````

## Implementation Tip: Staged Backpropagation

It is useful to break computations into **stages**.

Example:

Forward:
$$
\text{dot} = w \cdot x
$$

Backward:

* First compute gradient of **dot**
* Then propagate to ( w ) and ( x )

---

### Why This Matters

* Makes code **cleaner**
* Easier to debug
* Easier to extend

---

### Final Insight

Backpropagation design is flexible:

* You can **group operations**
* Or **split into smaller steps**

---

### Key Idea

> Choose a decomposition where:
>
> * Local gradients are simple
> * Implementation is efficient

---

### Summary

* Backpropagation = **local + chain rule**
* Gradients flow **backward**
* Computation graphs are **modular**
* Efficient implementations rely on **smart decomposition**

Consider the function:

$$
f(x, y) = \frac{x + \sigma(y)}{\sigma(x) + (x + y)^2}

> We do NOT need the full symbolic derivative

### Strategy: Break Into Steps

We decompose the function into **small intermediate computations**:

* Each step is simple
* Each step has a known local gradient

> Complex functions become manageable when broken into small steps.

Each intermediate variable:

* Stores a value
* Will later store a gradient

---

### Plan

1. Compute forward pass step-by-step
2. Store intermediate values
3. Run backward pass in reverse order
4. Apply chain rule at every step


````python
import math
x = 3 # example values
y = -4

# forward pass
sigy = 1.0 / (1 + math.exp(-y)) # sigmoid in numerator   #(1)
num = x + sigy # numerator                               #(2)
sigx = 1.0 / (1 + math.exp(-x)) # sigmoid in denominator #(3)
xpy = x + y                                              #(4)
xpysqr = xpy**2                                          #(5)
den = sigx + xpysqr # denominator                        #(6)
invden = 1.0 / den                                       #(7)
f = num * invden # done!                                 #(8)
````

## Important Practical Notes

### Cache Forward Variables

To compute backprop efficiently:

* Store intermediate variables from forward pass:

  * sigy, num, sigx, xpy, xpysqr, den, invden

Without caching:

* You would need to recompute them (inefficient)

---

### Gradients Add Up at Forks

Notice:

* ( x ) and ( y ) are used multiple times
* Gradients flow back through multiple paths

Therefore we use:

```python
dx += ...
dy += ...
```

instead of:

```python
dx = ...
```

---

### Why?

This follows the **multivariable chain rule**:

> If a variable influences the output through multiple paths,
> the gradients from each path **must be added**

---

### Key Insight

> Backpropagation = accumulation of gradient signals from all paths

---

### Summary

* Break computation into small steps
* Cache intermediate values
* Apply chain rule locally
* Accumulate gradients when paths merge

---

### Final Takeaway

> You don’t need symbolic derivatives —
> just a structured way to compute gradients step by step.

## Behavior of Common Gates

### Add Gate

The add gate:

$$
f(x, y) = x + y
$$

Backward behavior:

* Distributes gradient **equally** to all inputs
* Independent of input values

Reason:

$$
\frac{\partial f}{\partial x} = 1, \quad \frac{\partial f}{\partial y} = 1
$$

So:

$$
\text{input gradients} = \text{output gradient}
$$

---

### Key Insight

> Add gate = **distribute gradient unchanged**

---

### Max Gate

The max gate:

$$
f(x, y) = \max(x, y)
$$

Backward behavior:

* Sends gradient to **only one input**
* The input with the **maximum value**

Reason:

* Gradient = 1 for max input
* Gradient = 0 for others

---

### Key Insight

> Max gate = **route gradient to the winner**

---

### Multiply Gate

The multiply gate:

$$
f(x, y) = xy
$$

Backward behavior:

$$
\frac{\partial f}{\partial x} = y, \quad \frac{\partial f}{\partial y} = x
$$

So:

* Each input receives:

  * The **other input value**
  * Multiplied by upstream gradient

---

### Key Insight

> Multiply gate = **swap inputs and scale gradients**

---

## Intuition Summary

* Add → distribute
* Max → select
* Multiply → scale and swap

These simple patterns form the foundation of backpropagation.


## Intuition from Example

Consider a circuit:

* Add gate passes gradient unchanged to all inputs
* Max gate selects one input
* Multiply gate scales gradients based on input values

Example observations:

* Add gate sends gradient **equally**
* Max gate sends gradient **only to largest input**
* Multiply gate produces **asymmetric gradients**

---

### Example Insight

If gradient at output is:

$$
2.00
$$

Then:

* Add gate → both inputs get 2.00
* Max gate → only one input gets 2.00
* Multiply gate → gradients depend on input values

---

### Important Observation

Multiply gate behavior depends on **input magnitude**.

This leads to interesting effects.


## Unintuitive Effects and Consequences

### Gradient Imbalance in Multiply Gate

If:

* One input is **very large**
* Other input is **very small**

Then:

* Small input → gets **large gradient**
* Large input → gets **small gradient**

---

### Why This Happens

Because:

$$
\frac{\partial f}{\partial x} = y
$$

$$
\frac{\partial f}{\partial y} = x
$$

So gradients depend directly on the **other input’s value**.

---

### Practical Consequence

In linear models:

$$
w^T x
$$

If input data ( x ) is scaled:

* Example: multiply all inputs by 1000

Then:

* Gradients on weights also scale by 1000

---

### Implication

You must adjust learning rate:

* Larger inputs → larger gradients
* Require **smaller learning rate**

---

### Why This Matters

> Data preprocessing directly affects gradient behavior.

---

### Key Takeaway

* Gradient flow depends on:

  * Operation type
  * Input magnitude

---

### Final Insight

> Understanding gradient flow helps:
>
> * Debug training issues
> * Choose correct preprocessing
> * Set proper learning rates


## Gradients for Vectorized Operations

### Motivation

So far, we focused on **scalar variables**.

However, in practice:

* Neural networks operate on **vectors and matrices**
* All concepts extend naturally

---

### Key Difference

When working with vectors/matrices:

* Must pay attention to:

  * **Dimensions**
  * **Transpose operations**

---

### Important Insight

> Backpropagation rules stay the same —
> only the **shapes and linear algebra** become more important.

---

## Matrix-Matrix Multiplication

Consider:

$$
D = W \cdot X
$$

Where:

* $$ (W \in \mathbb{R}^{5 \times 10}) $$
* $$ (X \in \mathbb{R}^{10 \times 3}) $$
* $$ (D \in \mathbb{R}^{5 \times 3})$$


---

### Goal

Given:

$$
\frac{\partial L}{\partial D}
$$

We want:

$$
\frac{\partial L}{\partial W}, \quad \frac{\partial L}{\partial X}
$$

---

### Key Idea

Backprop through matrix multiplication follows the same principle:

> Apply chain rule + respect matrix dimensions

---


### Result (Important Formulas)

$$
dW = dD \cdot X^T
$$

$$
dX = W^T \cdot dD
$$

---

### Key Insight

* Gradient wrt weights → uses **input transpose**
* Gradient wrt input → uses **weights transpose**

---

### Intuition

This is just a **vectorized version of scalar multiplication rule**:

$$
f(x, y) = xy
\Rightarrow
\frac{\partial f}{\partial x} = y
\Rightarrow
\frac{\partial f}{\partial y} = x
$$

---

### Takeaway

> Matrix backprop = same rule, just with transposes
